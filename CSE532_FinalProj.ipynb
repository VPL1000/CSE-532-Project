## Part 5 — Linear Regression

Use regression to answer a predictive question and explain:
- Inputs (X)
- Output (y)
- Interpretation of results

```python

# TODO:
# Use:
# x = yearly_avg['year']
# y = yearly_avg['Temp']
#
# Run stats.linregress(x, y) and store the result in reg_result
#
# Print:
# - slope
# - intercept
# - r-value
# - p-value
# - standard error
"""
print(yearly_avg.head())
print(yearly_avg.columns)
print(yearly_avg.index)
"""
yearly_avg['year'] = pd.to_numeric(yearly_avg['year'], errors='coerce')
yearly_avg['Temp'] = pd.to_numeric(yearly_avg['Temp'], errors='coerce')

yearly_avg = yearly_avg.dropna(subset=['year', 'Temp'])

x = yearly_avg['year']
y = yearly_avg['Temp']

reg_result = stats.linregress(x, y)

print("Slope: ", reg_result.slope)
print("Intercept: ", reg_result.intercept)
print("R-Value: ", reg_result.rvalue)
print("P-Value: ", reg_result.pvalue)
print("Standard Error: ", reg_result.stderr)
```

```python

# TODO:
# Create a scatter plot of yearly average temperature vs year
# Add the regression line on the same plot
# Include:
# - title
# - x-label
# - y-label
# - legend


plt.figure(figsize=(8,5))

plt.scatter(yearly_avg['year'], yearly_avg['Temp'], label='Average Temp')

plt.plot(yearly_avg['year'], reg_result.intercept + reg_result.slope * yearly_avg['year'], label='Regression line')

plt.title('Yearly Average Temperature vs Year')
plt.xlabel('Year')
plt.ylabel('Temperature')
plt.legend()

plt.show
```

```python

# TODO:
# Create a predicted column for yearly_avg using the regression line
# Create a residual column:
# residual = actual - predicted
# Display the first 10 rows
'''
reg_result.slope
reg_result.intercept

predicted = (reg_result.slope * yearly_avg['year']) + reg_result.intercept
'''

yearly_avg['predicted'] = (reg_result.slope * yearly_avg['year']) + reg_result.intercept

yearly_avg['residual'] = yearly_avg['Temp'] - yearly_avg['predicted']


yearly_avg.head(10)
```

```python

# TODO:
# Plot residuals against year
# Add a horizontal line at y = 0
# Include:
# - title
# - x-label
# - y-label


plt.figure(figsize=(8,5))

plt.scatter(yearly_avg['year'], yearly_avg['residual'], label='Residual Temp')
plt.axhline(y=0, color='y')

plt.title('Residual Temperature between Observed and Expected vs Year')
plt.xlabel('Year')
plt.ylabel('Temperature Difference')
plt.legend()


plt.show
```
